# 2. Load Data

```python
data = pd.read_csv("../../data_sets/fetal_health.csv") #import
data.head() #peek
```

```python
data.info() #general overview of types and features
```

```python
print(data.shape) ## see shape of dataset

print("\nNum labels:")# number of labels
num_of_labels = len(data.fetal_health.unique()) 
print(num_of_labels)

data.describe() # see some stats
```

```python
# split train input and output data
X = data.drop(axis=0, columns=['fetal_health'])
Y = data.fetal_health # output = classifiy the fetal health (1, 2, 3)

#Print the shape of X and Y
print(X.shape)
print(Y.shape)
```

```python
# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```
