```python
#Fetal Health 
#Random Forest 

import pandas as pd 
import matplotlib.pyplot as plt #graphing 
import numpy as np 
import seaborn as sns 
pd.set_option ('display.max_columns',None) #this shows all of the rows 
```

```python
health = pd.read_csv('../../data_sets/fetal_health.csv') #this part of the code loads the data from the path 
health.head() #the health function is used to get the first n-th rows 
```

```python
health.info()  #the general overview of the data set "Health"
```

```python
health.describe() #compute a summary of statistics pertaining to the DataFrame  
```

```python
#Duplicate Data Check 

healthDd = health.copy() #this function determines if there's a duplicate in the dataset 
healthDd.drop_duplicates(inplace=True) #this function get rid of the duplicate if there are any in the dataset 
healthDd.shape #this function returns the shape of the array 
```

```python
health.shape #the orginal shape of the array before checking dupilcates 
```

```python
health=healthDd #replacing health data to healthDd data 
del healthDd #deleting healthDd  
health.head()
```

```python
#Exploratory Data Analysis (EDA) 

import pandas_profiling as pp 
profile = pp.ProfileReport(health) 
profile.to_file("EDA.html")
```

```python
X = health.drop('fetal_health', axis=1) #removes this specific row/column
y = health['fetal_health']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = 0.70, stratify=y, shuffle = True, random_state = 42)
```

```python
X_train.info()
```

```python
from sklearn.ensemble import RandomForestClassifier 
model = RandomForestClassifier(random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
#Score 

import sklearn.metrics 
score_cbc = model.score(X_test,y_test)
print("Score:", score_cbc) #accuracy score 
```

```python
#Confusion Matrix 

sklearn.metrics.confusion_matrix(y_test,model.predict(X_test))
```
